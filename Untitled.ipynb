```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample Data (replace with your own)
#np.random.seed(42)
dates = pd.to_datetime(['2024-01-01'] + [f'2024-01-{i:02d}' for i in range(2, 32)])
values = 10 + np.cumsum(np.random.randn(31))
df = pd.DataFrame({'Date': dates, 'Value': values})

df = pd.read_csv('data/performance_data.csv')

def plot_smoothed_trend(df, x_col, y_col, method='lowess', frac=0.1, window=5, order=3, show_original=True, **kwargs):
    """
    Plots a smoothed trend line.

    Args:
        df: Pandas DataFrame containing the data.
        x_col: Name of the column containing x-axis values (e.g., 'Date').
        y_col: Name of the column containing y-axis values (e.g., 'Value').
        method: Smoothing method ('lowess', 'rolling', 'polynomial').
        frac: Fraction of data used for LOWESS smoothing (0 < frac < 1). Only applicable for `method='lowess'`.
        window: Window size for rolling mean smoothing. Only applicable for `method='rolling'`.
        order: Order of the polynomial for polynomial smoothing.  Only applicable for `method='polynomial'`.
        show_original: Whether to plot the original data points.
        **kwargs: Additional keyword arguments passed to `sns.lineplot`.  For example, `color`, `linewidth`, etc.

    Returns:
        Matplotlib Axes object.
    """


    if show_original:  # Plot original data first for better visualization
        sns.scatterplot(data=df, x=x_col, y=y_col, color='lightgray', alpha=0.7)


    #if method == 'lowess':
    #    smoothed_values = sm.nonparametric.lowess(df[y_col], df[x_col], frac=frac, return_sorted=False)


    if method == "rolling":  # Moving/Rolling average
        print('here')
        smoothed_values = df[y_col].rolling(window=window, center=True, min_periods=1).mean()

    elif method == 'polynomial':
        x_numeric = pd.to_numeric(df[x_col])
        coeffs = np.polyfit(x_numeric, df[y_col], deg=order) # Fit polynomial
        smoothed_values = np.polyval(coeffs, x_numeric) # Evaluate polynomial at x points



    else:
        raise ValueError("Invalid smoothing method. Choose 'lowess', 'rolling', or 'polynomial'.")


    ax = sns.lineplot(x=df[x_col], y=smoothed_values, **kwargs)  # Plot smoothed line
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.title(f"Smoothed Trend ({method})")
    return ax



# Example usage:


# 1. LOWESS smoothing (Local Regression)
#import statsmodels.api as sm # Make sure you have statsmodels installed: `pip install statsmodels`

#plot_smoothed_trend(df, 'Date', 'Value', method='lowess', frac=0.2, color='blue', linewidth=2)
#plt.show()




# 2. Rolling Average Smoothing
plot_smoothed_trend(df, 'Step', 'Value', method='rolling', window=20, color='green', linewidth=2)
plt.show()



# 3. Polynomial Smoothing (requires numerical x-values or conversion, as shown here)
plot_smoothed_trend(df, 'Step', 'Value', method='polynomial', order=40, color='red', linewidth=2)
plt.show()
```
